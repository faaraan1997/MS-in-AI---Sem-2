```python
import json
from pprint import pprint
import nltk
nltk.download('stopwords')
from Questgen import main


with open('transcript.json', 'r') as file:
    data = json.load(file)

# Create the BoolQGen object
qe = main.BoolQGen()

# Iterate over each entry in the JSON file
for entry in data:
    # Extract the transcript text from the entry
    transcript_text = entry['transcript']

    payload = {"input_text": transcript_text}
    
    # Generate output based on the payload
    output = qe.predict_boolq(payload)
    
    # Print the output
    pprint(output)
```
